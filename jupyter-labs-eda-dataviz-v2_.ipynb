## Exploratory Data Analysis


First, let's read the SpaceX dataset into a Pandas dataframe and print its summary


```python
# from js import fetch
# import io

URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv"
# resp = await fetch(URL)
# dataset_part_2_csv = io.BytesIO((await resp.arrayBuffer()).to_py())
df=pd.read_csv(URL)
df.head(5)
```

All launch site names

First, let's try to see how the `FlightNumber` (indicating the continuous launch attempts.) and `Payload` variables would affect the launch outcome.

We can plot out the <code>FlightNumber</code> vs. <code>PayloadMass</code> and overlay the outcome of the launch. We see that as the flight number increases, the first stage is more likely to land successfully. The payload mass also appears to be a factor; even with more massive payloads, the first stage often returns successfully.


```python
sns.catplot(y="PayloadMass", x="FlightNumber", hue="Class", data=df, aspect = 4)
plt.xlabel("Flight Number",fontsize=20)
plt.ylabel("Pay load Mass (kg)",fontsize=20)
plt.show()
```

The scatter plot for <code>FlightNumber</code> vs. <code>PayloadMass</code>, with an overlay of the outcome of the launch. It was found that as the flight number increases, the first stage is more likely to land successfully. The payload mass also appears to be a factor; even with more massive payloads, the first stage often returns successfully.

### TASK 1: Visualize the relationship between Flight Number and Launch Site


Use the function <code>catplot</code> to plot <code>FlightNumber</code> vs <code>LaunchSite</code>, set the  parameter <code>x</code>  parameter to <code>FlightNumber</code>,set the  <code>y</code> to <code>Launch Site</code> and set the parameter <code>hue</code> to <code>'class'</code>


```python
# Plot a scatter point chart with x axis to be Flight Number and y axis to be the launch site, and hue to be the class value
sns.catplot(data=df, y='LaunchSite', x='FlightNumber', hue='Class', aspect= 3)
plt.xlabel('Flight number', fontsize=20)
plt.ylabel('Launch site', fontsize=20)
plt.show()
 
```

Explaining the patterns found in the <code>FlightNumber</code> vs. <code>LaunchSite</code>, with an overlay of the launch outcomes <code>(Class)</code>.
* It was found that as the flight number increased, the first stage was more likely to land successfully, for all the launch sites observed. 

### TASK 2: Visualize the relationship between Payload Mass and Launch Site


We also want to observe if there is any relationship between launch sites and their payload mass.


```python
# Plot a scatter point chart with x axis to be Pay Load Mass (kg) and y axis to be the launch site, and hue to be the class value
sns.catplot(data=df, y='LaunchSite', x='PayloadMass', hue='Class', aspect=3.0)
plt.xlabel('Payload mass (kg)', fontsize=20)
plt.ylabel('Launch site', fontsize=20)
plt.show()
```

Explaining the patterns found in the <code>PyloadMass</code> vs. <code>LaunchSite</code> scatter plot, with an overlay of the launch outcomes <code>(Class)</code>.
* It was found that, for the <code>VAFB-SLC 4E</code> launch site, there were no rockets launched for heavy payload mass greater than 10000 kg. 

* Yet both the launch sites <code>CCAFS-SLC 4E</code> and <code>KSC-LC 4E</code> were favoured by a high proportion of successful launches for heavy payload mass exceeding 10 000 kg.

### TASK  3: Visualize the relationship between success rate of each orbit type


Next, we want to visually check if there are any relationship between success rate and orbit type.


Let's create a `bar chart` for the sucess rate of each orbit


```python
print(df['Orbit'].unique())
df_group1 = df[['Orbit', 'Class']]
df_group1 = df_group1.groupby(by=['Orbit']).mean().reset_index()

plt.figure(figsize=(7, 3))
sns.barplot(data=df_group1, y='Class', x='Orbit')
plt.title('Barplot of Class by Orbit')
plt.xlabel('Orbit type')
plt.ylabel('Launch success rate')
plt.show()
```

Explaining the relationship found in the bar chart for the <code>Orbit</code> type vs. launch success rate <code>(Class)</code>.
* The bar chart shows that four orbit types: <code>ES-L1</code>, <code>GEO</code>, <code>HEO</code>, and <code>SSO</code>, associated with the highest  
launch success rate.

### TASK  4: Visualize the relationship between FlightNumber and Orbit type


For each orbit, we want to see if there is any relationship between FlightNumber and Orbit type.


```python
# Plot a scatter point chart with x axis to be FlightNumber and y axis to be the Orbit, and hue to be the class value
sns.catplot(data=df, y='Orbit', x='FlightNumber', hue='Class', aspect=3)
plt.xlabel('Number of flights', fontsize=20)
plt.ylabel('Orbit type', fontsize=20)
plt.show()
```

Explaining the scatter plot relationship of <code>Orbit</code> type vs. number of flights <code>(FlightNumber)</code>, with an overlay of the outcome of the launch.
* It can be seen that in the <code>LEO</code> orbit, launch success appears to be directly related to the number of flights. Conversely, in the <code>GTO</code> orbit, there appears to be no relationship between flight number and launch success. 

### TASK  5: Visualize the relationship between Payload Mass and Orbit type


Similarly, we can plot the Payload Mass vs. Orbit scatter point charts to reveal the relationship between Payload Mass and Orbit type


```python
# Plot a scatter point chart with x axis to be Payload Mass and y axis to be the Orbit, and hue to be the class value
sns.catplot(data=df, y='Orbit', x='PayloadMass', hue='Class', aspect=3)
plt.xlabel('Payload mass (kg)', fontsize=20)
plt.ylabel('Orbit type', fontsize=20)
plt.show()
```

Explaining the scatter plot relationship of <code>Orbit</code> type vs. <code>PayloadMass</code>, with an overlay of the outcome of the launch <code>(Class)</code>.
* With increasing payloads, there is a high distribution of successful landing rate for <code>LEO</code>, <code>ISS</code>, <code>PO</code> and <code>VLEO</code> orbits.  

* However, for <code>GTO</code> orbit, it's difficult to distinguish between successful and unsuccessful landings as both outcomes are present.

### TASK  6: Visualize the launch success yearly trend


You can plot a line chart with x axis to be <code>Year</code> and y axis to be average success rate, to get the average launch success trend.


The function will help you get the year from the date:


```python
# A function to Extract years from the date 
# year=[]
# def Extract_year():
#     for i in df["Date"]:
#         year.append(i.split("-")[0])
#     return year
# Extract_year()
# df['Date'] = year
# df.head()

year = [] 
def Extract_year():
    for i in df['Date']:
        year.append(i.split('-')[0])
    return year
# call the function
Extract_year()
df['Date'] = year
df.head()
```

```python
# Plot a line chart with x axis to be the extracted year and y axis to be the success rate

# organise data 
df_year = df[['Date', 'Class']]
df_year = df_year.groupby(by='Date').mean().reset_index()
x_data = df_year['Date']
y_data = df_year['Class']
# visulise data as a line plot
plt.figure(figsize=(8, 4))
plt.plot(x_data, y_data)
plt.xlabel('Year')
plt.ylabel('Launch success rate')
plt.show()
```

Explaining the line graph relationship of the launch success rate <code>(Class)</code> vs. time <code>(Year)</code>.
* The graph shows that the launch success rate experienced an increase from 2013 until 2020.
