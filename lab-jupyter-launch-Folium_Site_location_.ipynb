Next, let's try to enhance the map by adding the launch outcomes for each site, and see which sites have high success rates.

Next, let's create markers for all launch records.
If a launch was successful `(class=1)`, then we use a green marker and if a launch was failed, we use a red marker `(class=0)`


Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.


Let's first create a `MarkerCluster` object


```python
marker_cluster = MarkerCluster()
```

*TODO:* Create a new column in `spacex_df` dataframe called `marker_color` to store the marker colors based on the `class` value


```python

# Create a function to check the value of `class` column
# If class=1, marker_color value will be green
# If class=0, marker_color value will be red

def assign_marker_color(class_value):
    if class_value == 1:
        return 'green'
    else:
        return 'red'

# Then apply the function to create the 'marker_color' column in spacex_df
spacex_df['marker_color'] = spacex_df['class'].apply(assign_marker_color)
spacex_df.head()
```

*TODO:* For each launch result in `spacex_df` data frame, add a `folium.Marker` to `marker_cluster`


```python
# Add marker_cluster to current site_map
site_map.add_child(marker_cluster)

# For each row in spacex_df data frame, create a Marker object with its coordinate and customize the Marker's icon property
for index, row in spacex_df.iterrows():
    # Create and add a Marker cluster to the site map
    marker = folium.Marker(
        location=[row['Lat'], row['Long']],
        popup=row['Launch Site'],
        icon=folium.Icon(
            color='white',
            icon_color=row['marker_color'],
            html='<div style="font-size: 12px; color: #d35400;"><b>%s</b></div>' % row['Launch Site'],
        )
    ).add_child(folium.Popup(row['Launch Site']))
    marker_cluster.add_child(marker)

# Save and display the map
site_map.save('launch_sites_map_with_markers.html')
site_map
```

Next, we need to explore and analyze the proximities of launch sites.


Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
```

Now zoom in to a launch site and explore its proximity to see if you can easily find any railway, highway, coastline, etc. Move your mouse to these points and mark down their coordinates (shown on the top-left) in order to the distance to the launch site.


```python
# use the Haversine formular
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of the earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

*TODO:* Mark down a point on the closest coastline using MousePosition and calculate the distance between the coastline point and the launch site.


```python
# find coordinate of the closet coastline
# e.g.,: Lat: 28.56367  Lon: -80.57163
# distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
launch_site_coords = [34.632834, -120.610745]   # the chosen launch site: KSC LC-39A 
coastline_coords = [34.63247, -120.62718]
launch_site_lat = 34.632834
launch_site_lon = -120.610745
coastline_lat = 34.63247
coastline_lon = -120.62718
                 
distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
distance_text = f'Distance: {distance_coastline:.2f} km'
```

```python
# Create and add a folium.Marker on your selected closest coastline point on the map
# Display the distance between coastline point and launch site using the icon property 
# for example
distance_marker = folium.Marker(
   coastline_coords,
   icon=DivIcon(
       icon_size=(20,20),
       icon_anchor=(0,0),
       html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} km".format(distance_coastline),
       )
   )
site_map.add_child(distance_marker)
```

```python
# make each launch site with its list of coordinates
launchSites_names = ['CCAFS LC-40', 'CCAFS SLC-40', 'KSC LC-39A', 'VAFB SLC-4E'] # launch site 1 - 4
launchSites_df_T = launch_sites_df.T            # transpose the launch sites dataframe
launchSites_df_T.columns = launchSites_df_T.iloc[0]
launchSites_df_T = launchSites_df_T.drop(launchSites_df_T.index[0])
# separate launch site coordinates
lauchSite1_coordinates = launchSites_df_T['CCAFS LC-40'].to_list()      # launch site 1
lauchSite2_coordinates = launchSites_df_T['CCAFS SLC-40'].to_list()     # launch site 2
lauchSite3_coordinates = launchSites_df_T['KSC LC-39A'].to_list()       # launch site 3
lauchSite4_coordinates = launchSites_df_T['VAFB SLC-4E'].to_list()      # launch site 4

print(launchSites_df_T)
```

```python
# Launch site coordinates with unique proximity features respectively
  
launch_sites_features = {
# launch site CCAFS LC-40      
    'Launch Site 1': {  
        'coordinates': lauchSite1_coordinates,      # use MousePosition  
        'proximities': {  
            'City': [28.6115, -80.80837],           
            'Railway': [28.56228, -80.58713],  
            'Highway': [28.56229, -80.57062],  
            'Coastline': [28.56233, -80.56761]  
        }  
    },
# launch site KSC LC-39A
    'Launch Site 2': {  
        'coordinates': lauchSite2_coordinates,      # use MousePosition
        'proximities': {  
            'City': [28.6115, -80.80837],  
            'Railway': [28.56318, -80.58696],  
            'Highway': [28.5632, -80.57075],  
            'Coastline': [28.56311, -80.56781]         
        }  
    }, 
# launch site CCAFS SLC-40
    'Launch Site 3': {  
        'coordinates': lauchSite3_coordinates,      # use MousePosition  
        'proximities': {  
            'City': [28.6115, -80.80837],  
            'Railway': [28.5732, -80.65394],  
            'Highway': [28.57313, -80.65544],  
            'Coastline': [28.57307, -80.57002]   
        }  
    },  
# launch site VAFB SLC-4E
    'Launch Site 4': {  
        'coordinates': lauchSite4_coordinates,  
        'proximities': {  
            'City': [34.95237, -120.43619],  
            'Railway': [34.63363, -120.62437],  
            'Highway': [34.63180, -120.18714],  
            'Coastline': [34.63285, -120.62640]   
        }  
    }  
}  

# Iterate through launch sites  
for site_name, site_data in launch_sites_features.items():  
    site_coords = site_data['coordinates']  
    
    # Add launch site marker  
    launchSite_marker = folium.Marker(  
        site_coords,   
        popup=site_name,  
        icon=folium.Icon(color='red', icon='rocket')  
    ).add_to(site_map)  
    
    # Iterate through site-specific proximities  
    for feature_name, feature_coords in site_data['proximities'].items():  
        # Calculate distance  
        distance = calculate_distance(  
            site_coords[0], site_coords[1],  
            feature_coords[0], feature_coords[1]  
        )
          
        
        # Proximity feature marker  
        proximity_marker = folium.Marker(  
            feature_coords,  
            icon=folium.Icon(color='blue', icon='info-sign')  
        ).add_to(site_map)  
        
        # Distance marker  
        distance_marker = folium.Marker(  
            feature_coords,  
            icon=DivIcon(  
                icon_size=(200, 36),  
                icon_anchor=(0, 0),  
                html=f'<div style="font-size: 12px; color:#d35400;">'  
                     f'<b>{site_name} - {feature_name}: {distance:.2f} km</b></div>'  
            )  
        ).add_to(site_map)  
        
        # Line connecting launch site and feature  
        lines = folium.PolyLine(  
            locations=[site_coords, feature_coords],  
            color='green',  
            weight=2,  
            opacity=0.7,  
            dash_array='5'  
        ).add_to(site_map)  

# Display the map
site_map.add_child(launchSite_marker)
site_map.add_child(proximity_marker)  
site_map.add_child(distance_marker)
site_map.add_child(lines)

site_map
```
